# $y=bx$

$y=bx+\epsilon$

$y|x \sim N(bx,\sigma^2)$

$x \sim N(\mu_0,\sigma_0^2)$

Then

$x|y \sim N(\mu_p,\sigma_p^2)$

where

$\mu_p = \frac{\sigma_0^2by+\sigma^2\mu_0}{\sigma_0^2b^2+\sigma^2}$

and

$\sigma_p^2 = \frac{2\sigma_0^2\sigma^2}{\sigma_0^2b^2+\sigma^2}$

```python
p=3

rl = -10
ru=20

obs_error = 0.001

x=torch.linspace(rl,ru,p)

b=0.5

def lin(x):
    y = b*x
    return y

y = lin(x) 

plt.plot(x.detach().numpy(),y.detach().numpy(),'o')

tp = 500
dx = (ru-rl)/tp
x_test = torch.linspace(ru,rl,tp)
```

```python
dist_prior = normal.Normal(loc=0, scale=1)
val_prior = torch.exp(dist_prior.log_prob(x_test))
```

```python
def acceptance(y,x,x_prime,sigma2): # Log acceptance for stability
    ratio = (-0.5*x_prime**2-0.5*(1/sigma2)*(b*x_prime-y)**2)-(-0.5*x**2-0.5*(1/sigma2)*(b*x-y)**2)
    alpha = np.min((0,ratio))
    return alpha

```

```python
def metrop(y_cal,x0,sigma2_mc,sigma2_e,iters):
    chain = np.zeros(iters+1) #Initialise markov chain
    chain[0]=x0 #set x0 as first element
    for i in range(iters):
        x = chain[i] #define x
        x_p = np.random.normal(chain[i],sigma2_mc) #select candidate value x'
        alpha = acceptance(y,x,x_p,sigma2_e) #calculate alpha
        u=np.random.uniform() #generate random uniform variable
        if np.log(u)<alpha: 
            chain[i+1]=x_p #accept candidate value
        else:
            chain[i+1]=x #reject candidate value
    return chain

```

```python
y=3
x0 = 0
sigma2_mc =1
sigma2_e=0.1
iters=10000
```

```python
posterior_mean = (1/((b**2)+sigma2_e))*b*y
posterior_cov = sigma2_e/(b**2 + sigma2_e)
    
dist_posterior = normal.Normal(loc=posterior_mean, scale=np.sqrt(posterior_cov))
posterior_True = torch.exp(dist_posterior.log_prob(x_test))
```

```python
X = metrop(y,x0,sigma2_mc,sigma2_e,iters)
```

```python
import emcee
ndim = 1
nwalkers = 2
p0 = np.random.normal(0,1,2)
y = 3
```

```python
def log_prior(theta):
    lp = -0.5*theta**2
    return lp

def log_likelihood(y,theta,sigma2_e):
    ll = -(0.5/(sigma2_e))*(b*theta-y)**2
    return ll
    
def log_prob(theta,y_val,sigma2_e):
    
    lp = log_prior(theta)
    if not np.isfinite(lp):
        return -np.inf
    
    return lp + log_likelihood(theta, y,sigma2_e)
```

```python
sampler = emcee.EnsembleSampler(nwalkers, ndim, log_prob, args=[y,sigma2_e])
```

```python
sampler.run_mcmc(p0[:,None], 10000)
```

```python
flat_samples = sampler.get_chain(discard=2000, thin=10, flat=True)
print(flat_samples.shape)
```
